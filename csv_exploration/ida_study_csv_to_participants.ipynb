### creation of participants.tsv

```python
# -------- CONFIG --------
participants_tsv = ida_csv.parent / "participants.tsv"

# -------- MANDATORY BIDS FIELDS --------
df_bids = pd.DataFrame()
df_bids["participant_id"] = df[id_column_name].apply(lambda x: f"sub-{x}")
df_bids["sex"] = df[sex_column_name].map(sex_map).fillna("U")
df_bids["age"] = df[age_column_name]

# -------- AUTOMATICALLY ADD ALL OTHER COLUMNS --------
mandatory_cols = {id_column_name, sex_column_name, age_column_name}  # already used
optional_cols = [c for c in df.columns if c not in mandatory_cols]

for col in optional_cols:
    df_bids[col] = df[col]

# -------- SORT (optional) --------
df_bids = df_bids.sort_values("participant_id")

# -------- WRITE TSV --------
df_bids.to_csv(participants_tsv, sep="\t", index=False, na_rep="")


print(f"participants.tsv written to {participants_tsv}")
# print(df_bids.head())
```
